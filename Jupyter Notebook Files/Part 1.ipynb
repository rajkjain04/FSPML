### 1.2.2   <a class="anchor" id="1.2.2"></a>

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
from statsmodels.tsa.arima_model import ARIMA 
from statsmodels.tsa.ar_model import AR 
import copy 
import warnings
warnings.filterwarnings("ignore") 

snp = pd.read_csv('snp_500_2015_2019.csv')
snp.set_index(['Date'], inplace = True)
snp_close = snp['Close'].to_frame().apply(np.log)

snp_arma = copy.deepcopy(snp_close)
snp_arma.columns = ['True'] 
model = ARIMA(snp_arma, order = (1,0,0)).fit()
snp_arma['Res'] = ARIMA(snp_arma, order = (1,0,0)).fit().resid
snp_arma['Prediction'] = snp_arma['True'] - snp_arma['Res']

snp_close.plot(color = 'red')
snp_arma['Prediction'].plot(color='blue')

plt.title('S&P 500 Close Log Price and ARMA(1,0) model prediction', fontname="arial", fontsize=16, fontweight="bold")
plt.xlabel('Date', fontname="arial", fontsize=16)
plt.ylabel('Log Price', fontname="arial", fontsize=16)
plt.xticks(fontname="arial", fontsize=16)
plt.yticks(fontname="arial", fontsize=16)

plt.grid(which='major', linestyle='-', linewidth='0.5', color='#666666')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)

plt.legend(['Close','ARMA(1,0)']);
# plt.rc('font', family = 'arial')

fig = plt.gcf()
fig.set_size_inches(8, 6)

# plt.savefig('1_2_2a.pdf')
```

```python
model.params
```

```python
arma_rmse = np.sqrt(np.mean(snp_arma["Res"]**2))
print("Root Mean Squared Error: ", f'{arma_rmse:.4}')
print("Mean Absolute Error: ",  np.mean(abs(snp_arma['Res'])))
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
from statsmodels.tsa.api import VAR 

df = pd.read_csv('snp_allstocks_2015_2019.csv')
df = df.set_index('Date')

info = pd.read_csv('snp_info.csv')
info.drop(columns=info.columns[0], inplace = True)

eig_val = pd.DataFrame(columns=info['GICS Sector'].unique(), index=['Minimum Eigenvalue', 'Average Eigenvalue', 'Maximum Eigenvalue'])

for sector in info['GICS Sector'].unique():
    
    tickers = info.loc[info['GICS Sector']==sector]['Symbol'].tolist()
    stocks = df[tickers]
    stocks_ma = stocks.rolling(window=66).mean()
    stocks_detrended = stocks.sub(stocks_ma).dropna()
    model = VAR(stocks_detrended)
    results = model.fit(1)
    A = results.params[1:].values
    eigA, _ = np.linalg.eig(A)
    eig_val[sector]['Minimum Eigenvalue'] = np.min(abs(eigA))
    eig_val[sector]['Average Eigenvalue'] = np.mean(abs(eigA))
    eig_val[sector]['Maximum Eigenvalue'] = np.max(abs(eigA))

eig_val_final = eig_val.T.round(5)
eig_val_final = eig_val_final.astype(float).round(6)
# eig_val_final.to_csv('eigdf.csv')
```
