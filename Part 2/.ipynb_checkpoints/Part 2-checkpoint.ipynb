```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv(r'fsp_case_31_BSDfinal.csv', index_col = 0, header = [0,1])

df = df.replace(r'^\s+$', np.nan, regex=True)

average_returns = df['ret'].mean(axis=1)
average_returns.plot(color = 'blue')
plt.title('Daily Market Returns from 02/01/2017 to 31/12/2018', fontname="arial", fontsize=14, fontweight="bold")

plt.xlabel('Date', fontname="arial", fontsize=14)
plt.ylabel('Market Returns', fontname="arial", fontsize=14)
plt.xticks(fontname="arial", fontsize=10)
plt.yticks(fontname="arial", fontsize=10)

plt.grid(which='major', linestyle='-', linewidth='0.5', color='#666666')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)

fig = plt.gcf()
fig.set_size_inches(6, 4)

plt.savefig('2_4_1compare.pdf') # Saves it as a vectorized plot 
```

### 2.4.2

```python
import pandas as pd
import numpy as np
df = pd.read_csv(r'fsp_case_31_BSDfinal.csv',index_col=0,header=[0,1])

market_daily_return = df[['ret']].fillna(0).mean(axis=1).to_frame()
market_daily_return.columns = ['Mean Market Return']
market_daily_return.index = pd.to_datetime(market_daily_return.index, format='%Y-%m-%d')

r_e = df['ret'].fillna(0)
r_e.columns = r_e.columns.tolist()
r_e.index = pd.to_datetime(r_e.index, format='%Y-%m-%d')
r_m = market_daily_return

def cal_beta(df):
    betas = []
    for index in range (22,len(df)+1):
        cov_matrix = df.iloc[(index-22):index].cov()
        if cov_matrix.iloc[0,0] == 0:
            betas.append(0)
        else:
            temp_beta = (cov_matrix.iloc[0,1])/(cov_matrix.iloc[1,1])
            betas.append(temp_beta)
          
    return betas

def cal_all_beta(df, r_m):
    all_betas = []
    for companies in df:
        all_betas.append(cal_beta(pd.concat([df[companies], r_m], axis = 1)))
    
    return pd.DataFrame(all_betas, index = df.columns, columns = df.index[21:]).transpose()


all_beta = cal_all_beta(r_e,r_m)
```

```python
# all_beta.iloc[:,0:].plot(legend = False)

plt.plot(all_beta.iloc[:,0:])

plt.title('22-Day Sliding Window Betas for 157 Assets', fontname="arial", fontsize=14, fontweight="bold")

plt.xlabel('Date', fontname="arial", fontsize=12)
plt.ylabel('Beta', fontname="arial", fontsize=12)
plt.xticks(fontname="arial", fontsize=9)
plt.yticks(fontname="arial", fontsize=10)

plt.grid(which='major', linestyle='-', linewidth='0.5', color='#666666')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)

fig.set_size_inches(6, 4)
plt.savefig('2_4_2compare.pdf')  
```

### 2.4.3

```python
import pandas as pd 
import numpy as np
# import datetime as dt
import matplotlib.pyplot as plt
df = pd.read_csv(r'fsp_case_31_BSDfinal.csv', index_col = 0, header = [0,1])

df = df.replace(r'^\s+$', np.nan, regex=True)

daily_total_mcap = df['mcap'].sum(axis=1)
mcap_times_return = np.multiply(df['mcap'],df['ret']).sum(axis=1)
daily_weighted_return = mcap_times_return / daily_total_mcap
percentage_return = 100 * daily_weighted_return

percentage_return.plot(color = 'blue')

plt.title('Daily Cap-Weighted Market Returns from 02/01/2017 to 31/12/2018', fontname="arial", fontsize=12, fontweight="bold")

plt.xlabel('Date', fontname="arial", fontsize=12)
plt.ylabel('Market Returns (%)', fontname="arial", fontsize=12)
plt.xticks(fontname="arial", fontsize=10)
plt.yticks(fontname="arial", fontsize=10)

plt.grid(which='major', linestyle='-', linewidth='0.5', color='#666666')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)

fig.set_size_inches(6, 4)
plt.savefig('2_4_3a.pdf')  
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv(r'fsp_case_31_BSDfinal.csv',index_col=0,header=[0,1])

df = df.replace(r'^\s+$', np.nan, regex=True)

daily_total_mcap = df['mcap'].sum(axis=1)
mcap_times_return = np.multiply(df['mcap'],df['ret']).sum(axis=1)
daily_weighted_return = mcap_times_return / daily_total_mcap

r_e = df['ret'].fillna(0)
r_e.columns = r_e.columns.tolist()
r_e.index = pd.to_datetime(r_e.index, format='%Y-%m-%d')
r_m = daily_weighted_return

def cal_beta(df):
    betas = []
    for index in range (22,len(df)+1):
        cov_matrix = df.iloc[(index-22):index].cov()
        if cov_matrix.iloc[0,0] == 0:
            betas.append(0)
        else:
            temp_beta = (cov_matrix.iloc[0,1])/(cov_matrix.iloc[1,1])
            betas.append(temp_beta)
          
    return betas

def cal_all_beta(df, r_m):
    all_betas = []
    for companies in df:
        all_betas.append(cal_beta(pd.concat([df[companies], r_m], axis = 1)))
    
    return pd.DataFrame(all_betas, index = df.columns, columns = df.index[21:]).transpose()


all_beta = cal_all_beta(r_e,r_m)
```

```python
plt.plot(all_beta.iloc[:,0:])

plt.title('22-Day Sliding Window Cap-Weighted Betas for 157 Assets', fontname="arial", fontsize=12, fontweight="bold")

plt.xlabel('Date', fontname="arial", fontsize=12)
plt.ylabel('Beta', fontname="arial", fontsize=12)
plt.xticks(fontname="arial", fontsize=9)
plt.yticks(fontname="arial", fontsize=10)

plt.grid(which='major', linestyle='-', linewidth='0.5', color='#666666')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)

fig.set_size_inches(6, 4)
plt.savefig('2_4_4.pdf')  
```
